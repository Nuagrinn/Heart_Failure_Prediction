Есть явные тенденции.

```python
# Функция для отображения распределения признака и его связи с таргетом 

def plot_stats(feature,label_rotation=False,horizontal_layout=True):
    temp = heart_failure_df[feature].value_counts()
    df1 = pd.DataFrame({feature: temp.index,'Количество наблюдений': temp.values})

    # Расчет доли target=1 в категории
    cat_perc = heart_failure_df[[feature, 'DEATH_EVENT']].groupby([feature],as_index=False).mean()
    cat_perc.sort_values(by='DEATH_EVENT', ascending=False, inplace=True)
    
    if(horizontal_layout):
        fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(12,6))
    else:
        fig, (ax1, ax2) = plt.subplots(nrows=2, figsize=(12,14))
    sns.set_color_codes("pastel")
    s = sns.barplot(ax=ax1, x = feature, y="Количество наблюдений",data=df1)
    if(label_rotation):
        s.set_xticklabels(s.get_xticklabels(),rotation=90)
    
    s = sns.barplot(ax=ax2, x = feature, y='DEATH_EVENT', order=cat_perc[feature], data=cat_perc)
    if(label_rotation):
        s.set_xticklabels(s.get_xticklabels(),rotation=90)
    plt.ylabel('Доля смертей', fontsize=10)
    plt.tick_params(axis='both', which='major', labelsize=10)

    plt.show();
```

```python
plot_stats('high_blood_pressure')
```

```python
# Предварительно удалим бесполезные признаки 
tr_data = heart_failure_df.drop(['time','sex','diabetes', 'smoking'], axis=1)
```

```python
X = tr_data.drop('DEATH_EVENT',axis=1)
y = tr_data.DEATH_EVENT
```

```python
X.head()
```

```python
# Sequential Feature Selection. Увеличение размерности
# http://rasbt.github.io/mlxtend/user_guide/feature_selection/SequentialFeatureSelector/
from mlxtend.feature_selection import SequentialFeatureSelector as SFS

# базовая модель для выбора фичей
RF_clf = RandomForestClassifier()

sfs1 = SFS(RF_clf, 
           k_features=8, 
           forward=True, 
           floating=False, 
           verbose=2,
           scoring='roc_auc',
           cv=4)

sfs1 = sfs1.fit(X, y)
```

Будем использоваться все 8 фичей. Но ради интереса производем еще один отбор признаков сайкит-лернинговским способом.

```python
import matplotlib.pyplot as plt

from sklearn.model_selection import StratifiedKFold
from sklearn.feature_selection import RFECV
from sklearn.datasets import make_classification

# Build a classification task using 3 informative features
X, y = make_classification(n_samples=100, n_features=8, n_informative=6,
                           n_redundant=2, n_repeated=0, n_classes=2,
                           n_clusters_per_class=1, random_state=42)

# Create the RFE object and compute a cross-validated score.
RF_clf = RandomForestClassifier()
# The "accuracy" scoring is proportional to the number of correct
# classifications
rfecv = RFECV(estimator=RF_clf, step=1, cv=StratifiedKFold(3),
              scoring='roc_auc')
rfecv.fit(X, y)

print("Optimal number of features : %d" % rfecv.n_features_)

# Plot number of features VS. cross-validation scores
plt.figure()
plt.xlabel("Number of features selected")
plt.ylabel("Cross validation score (nb of correct classifications)")
plt.plot(range(1, len(rfecv.grid_scores_) + 1), rfecv.grid_scores_)
plt.show()
```

Результат тот же

### Моделирование и классификация 

```python
# Тестовая и тренировочная выборки 
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.33, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
logreg = LogisticRegression()
logreg.fit(X_train, y_train)
print('Accuracy of Logistic regression classifier on training set: {:.2f}'
     .format(logreg.score(X_train, y_train)))
print('Accuracy of Logistic regression classifier on test set: {:.2f}'
     .format(logreg.score(X_test, y_test)))
```

```python
# XGBoost

XGB_clf = xgb.XGBClassifier(n_estimators=500)
XGB_clf.fit(X_train, y_train)
print('Accuracy of XG Boost classifier on training set: {:.2f}'
     .format(XGB_clf.score(X_train, y_train)))
print('Accuracy of XG Boost classifier on test set: {:.2f}'
     .format(XGB_clf.score(X_test, y_test)))
```

```python
RF_clf = RandomForestClassifier(n_jobs=-1, random_state=42)

RF_clf.fit(X_train, y_train)
print('Accuracy of Random Forest classifier on training set: {:.2f}'
     .format(RF_clf.score(X_train, y_train)))
print('Accuracy of Random Forest classifier on test set: {:.2f}'
     .format(RF_clf.score(X_test, y_test)))
```

```python
# предсказываем классы и его вероятности 
y_pred = RF_clf.predict(X_test)
y_predicted_prob = RF_clf.predict_proba(X_test)
```

```python
# roc auc
fpr, tpr, thresholds = roc_curve(y_test, y_pred)
roc_auc= auc(fpr, tpr)

plt.figure()
plt.plot(fpr, tpr, color='darkorange', label='ROC curve (area = %0.2f)' % roc_auc)
plt.plot([0, 1], [0, 1], color='navy', linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver operating characteristic example')
plt.legend(loc="lower right")
plt.show
```

Базовый xgb-классификатор показал результат хуже, чем случайный лес. Попробуем настроить его и подобрать параметры.

```python
# Словарь с вариацией параметров из которых будем выбирать оптимальный набор
params = {
        'min_child_weight': [1, 5, 10],
        'gamma': [0.5, 1, 1.5, 2, 5],
        'subsample': [0.6, 0.8, 1.0],
        'colsample_bytree': [0.6, 0.8, 1.0],
        'max_depth': [3, 4, 5, 7, 9, 11]
        }
```

```python
# Базовая модель, для которой будем подбирать параметры

xgb_clf = xgb.XGBClassifier(learning_rate=0.02, n_estimators=1000, objective='binary:logistic',
                    silent=True, nthread=1)
```

```python
# Функция для подсчета времени. Будем приятным дополнением в следующем пункте.

def timer(start_time=None):
    if not start_time:
        start_time = datetime.now()
        return start_time
    elif start_time:
        thour, temp_sec = divmod((datetime.now() - start_time).total_seconds(), 3600)
        tmin, tsec = divmod(temp_sec, 60)
        print('\n Time taken: %i hours %i minutes and %s seconds.' % (thour, tmin, round(tsec, 2)))
```

```python
# Количество фолдов для кросс-валидации 
folds = 3
param_comb = 5

skf = StratifiedKFold(n_splits=folds, shuffle = True, random_state = 1001)

random_search = RandomizedSearchCV(xgb_clf, param_distributions=params, n_iter=param_comb, scoring='roc_auc', n_jobs=-1, cv=skf.split(X,y), verbose=3, random_state=1001 )

# Время на поиск оптимальных параметров
start_time = timer(None) 
random_search.fit(X, y)
timer(start_time) 
```

```python
XGB_best = random_search.best_estimator_
```

```python
# Предсказанные значения
y_pred=XGB_best.predict(X_test)

# Точность
accuracy=accuracy_score(y_pred, y_test)
print('Model accuracy score: {0:0.4f}'.format(accuracy_score(y_test, y_pred)))
```

```python
# ROC-AUC

y_predicted_prob = XGB_best.predict_proba(X_test)

from sklearn.metrics import roc_curve, auc
fpr, tpr, thresholds = roc_curve(y_test, y_predicted_prob[:,1])
roc_auc= auc(fpr, tpr)
plt.figure()
plt.plot(fpr, tpr, color='darkorange', label='ROC curve (area = %0.2f)' % roc_auc)
plt.plot([0, 1], [0, 1], color='navy', linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver operating characteristic example')
plt.legend(loc="lower right")
plt.show()
```
